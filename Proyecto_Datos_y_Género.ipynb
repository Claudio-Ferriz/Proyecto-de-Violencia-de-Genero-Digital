```python
# Se decide eliminar la columna '33. Si querés recibir el informe con los
# resultados de la encuesta, compartinos tu correo electrónico', ya que si
# tendrían datos de correos electrónicos, estos afectarían la anonimización de
# los datos y la privacidad de las personas encuestadas. A su vez estos datos no
# aportarían valor al análisis que se esta realizando.

df_sd = df_sd.drop(df_sd.columns[34], axis=1)
```

```python
# Se genera un array para que se puedan ver los distintos valores que componen
# la columna 'Provincia'.

df_sd['Provincia'].unique()
```

```python
# Se genera una nueva columna que agrupa a las distintas provincias según
# distintas regiones del país.

CABA = ['Ciudad Autónoma de Buenos Aires (CABA)']

Centro = ['Buenos Aires',  'Córdoba', 'Santa Fe', 'Entre Ríos', 'La Pampa']

Cuyo = ['San Luis', 'San Juan', 'Mendoza']

NEA = ['Misiones', 'Corrientes', 'Chaco', 'Formosa']

NOA = ['Santiago del Estero', 'Salta', 'Tucumán', 'Jujuy', 'Catamarca', 'La Rioja']

Patagonia = ['Santa Cruz', 'Tierra del Fuego', 'Río Negro', 'Chubut', 'Neuquén']

condiciones = [
    df_sd['Provincia'].isin(CABA),
    df_sd['Provincia'].isin(Centro),
    df_sd['Provincia'].isin(Cuyo),
    df_sd['Provincia'].isin(NEA),
    df_sd['Provincia'].isin(NOA),
    df_sd['Provincia'].isin(Patagonia)
]

df_sd['Región'] = np.select(condiciones,['CABA', 'Centro', 'Cuyo', 'NEA', 'NOA', 'Patagonia'], default= 'Otros')
```

```python
# Convertimos a numérico, los errores se vuelven NaN
df_sd['Edad'] = pd.to_numeric(df_sd['Edad'], errors='coerce')

# Opcional: Eliminar filas donde la edad sea nula ahora que es numérica
df_sd = df_sd.dropna(subset=['Edad'])

# Convertir a entero (int)
df_sd['Edad'] = df_sd['Edad'].astype(int)
```

```python
# Definimos los límites de los grupos
# 0-17 (Menores), 18-29 (Jóvenes), 30-45 (Adultos), 46-60 (Adultos mayores), 60+ (Seniors)
bins = [0, 17, 29, 45, 60, 110]
labels = ['0-17', '18-29', '30-45', '46-60', '60+']

# Creamos la nueva columna
df_sd['Segmento de edad'] = pd.cut(df_sd['Edad'], bins=bins, labels=labels, right=True)

```

```python
# Se genera un gráfico con los totales de situaciones de violencia de género en
# entornos digitales con conocimiento de los agresores discriminados
# por provincias, y a su vez incluyen las categorías de nivel de conocimiento
# hacia los agresores.

df_pivot_provincia = df_sd.groupby(['Provincia', 'Si existe conocimiento de la persona o grupo que llevo a cabo la acción']).size().unstack(fill_value=0)

df_pivot_provincia['Total'] = df_pivot_provincia.sum(axis=1)
df_pivot_provincia = df_pivot_provincia.sort_values('Total', ascending=True).drop(columns='Total')

ax = df_pivot_provincia.plot(kind='barh', stacked=True, figsize=(12, 8), colormap='viridis')

plt.xlabel('Cantidad total de situaciones de violencia de género con conocimiento de los agresores')
plt.ylabel('Provincias')
plt.title('Situaciones de violencia de género con conocimiento de los agresores segmentadas por provincia')
plt.legend(title='Nivel de conocimiento', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='x', linestyle='--', alpha=0.7)

for p in ax.patches:
    width = p.get_width()
    if width > 0:
        ax.annotate(str(int(width)),
                    (p.get_x() + width/2, p.get_y() + p.get_height()/2),
                    ha='center', va='center', fontsize=9, color='white')

plt.tight_layout()
plt.show()
```

```python
# Se genera el mismo gráfico que el anterior pero discriminado por regiones.

df_pivot_región = df_sd.groupby(['Región', 'Si existe conocimiento de la persona o grupo que llevo a cabo la acción']).size().unstack(fill_value=0)

df_pivot_región['Total'] = df_pivot_región.sum(axis=1)
df_pivot_región = df_pivot_región.sort_values('Total', ascending=True).drop(columns='Total')

ax = df_pivot_región.plot(kind='barh', stacked=True, figsize=(12, 8), colormap='viridis')

plt.xlabel('Cantidad total de situaciones de violencia de género con conocimiento de los agresores')
plt.ylabel('Regiones')
plt.title('Situaciones de violencia de género con conocimiento de los agresores segmentadas por región')
plt.legend(title='Nivel de conocimiento', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='x', linestyle='--', alpha=0.7)

for p in ax.patches:
    width = p.get_width()
    if width > 0:
        ax.annotate(str(int(width)),
                    (p.get_x() + width/2, p.get_y() + p.get_height()/2),
                    ha='center', va='center', fontsize=9, color='white')

plt.tight_layout()
plt.show()
```

```python
# 1. Preparar el DataFrame original (Cantidades absolutas)
df_pivot_región = df_sd.groupby(['Región', 'Si existe conocimiento de la persona o grupo que llevo a cabo la acción']).size().unstack(fill_value=0)

# Ordenar por el total para que el gráfico sea legible
df_pivot_región['Total'] = df_pivot_región.sum(axis=1)
df_pivot_región = df_pivot_región.sort_values('Total', ascending=True).drop(columns='Total')

# 2. Crear el DataFrame de porcentajes (Para el dibujo de las barras)
df_pivot_región_perc = df_pivot_región.div(df_pivot_región.sum(axis=1), axis=0) * 100

# 3. Graficar usando el DataFrame de porcentajes
ax = df_pivot_región_perc.plot(kind='barh', stacked=True, figsize=(12, 8), colormap='viridis')

plt.xlabel('Distribución porcentual (%)')
plt.ylabel('Regiones')
plt.title('Situaciones de violencia de género con conocimiento de agresores segmentadas por región (Porcentajes)')
plt.legend(title='Nivel de conocimiento', bbox_to_anchor=(1.05, 1), loc='upper left')

# 4. Anotar las CANTIDADES ABSOLUTAS sobre las barras porcentuales
# Usamos un contador para iterar sobre los datos originales
for i, (idx, row) in enumerate(df_pivot_región.iterrows()):
    cum_width = 0
    for col in df_pivot_región.columns:
        # El ancho visual en el gráfico es el porcentaje
        width_perc = df_pivot_región_perc.loc[idx, col]
        # El valor real que queremos mostrar es el conteo
        real_value = row[col]

        if real_value > 0:
            ax.annotate(str(int(real_value)),
                        (cum_width + width_perc/2, i),
                        ha='center', va='center',
                        fontsize=9, color='white', fontweight='bold')
        cum_width += width_perc

plt.xlim(0, 100)
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
